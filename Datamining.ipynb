```python
mean = df['Suhu_Udara'].mean()
median = df['Suhu_Udara'].median()
modus = df['Suhu_Udara'].mode()[0]
print('mean =', mean)
print('median =', median)
print('modus =', modus)
```

```python
mean = df['Curah_Hujan'].mean()
median = df['Curah_Hujan'].median()
modus = df['Curah_Hujan'].mode()[0]
print('mean =', mean)
print('median =', median)
print('modus =', modus)
```

```python
# Visualisasi dengan Box plot
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)

# Plot histogram untuk curah hujan
plt.figure(figsize=(8, 6))
plt.hist(df['Curah_Hujan'], bins=10, edgecolor='green')
plt.xlabel('Curah Hujan')
plt.ylabel('Frekuensi')
plt.title('Histogram Curah hujan')

# Tambahkan garis vertikal untuk menunjukkan nilai mean 
mean_curah_hujan = df['Curah_Hujan'].mean()
plt.axvline(mean_curah_hujan, color='red', linestyle='dashed', linewidth=1, label='mean')
plt.text(mean_curah_hujan+1, 20, f'mean: {mean_curah_hujan:.2f}', color='black')

# Tambahkan garis vertikal untuk menunjukkan nilai standar deviasi
std_curah_hujan = df['Curah_Hujan'].std()
plt.axvline(mean_curah_hujan+std_curah_hujan, color='blue', linestyle='dashed', linewidth=1, label='Std Dev')
plt.axvline(mean_curah_hujan+std_curah_hujan, color='blue', linestyle='dashed', linewidth=1)
plt.text(mean_curah_hujan+std_curah_hujan+1, 15, f'Std Dev: {std_curah_hujan:.2f}', color='blue')

plt.legend()
plt.show()

# Plot histogram untuk suhu
plt.figure(figsize=(8, 6))
plt.hist(df['Suhu_Udara'], bins=10, edgecolor='green')
plt.xlabel('Suhu')
plt.ylabel('Frekuensi')
plt.title('Histogram Suhu')
plt.legend()
plt.show()

# Plot histogram untuk suhu
plt.figure(figsize=(8, 6))
plt.hist(df['Suhu_Udara'], bins=10, edgecolor='green')
plt.xlabel('Suhu')
plt.ylabel('Frekuensi')
plt.title('Histrogram Suhu')

# Tambahkan garis vertikal untuk menunjukkan nilai mean
mean_suhu = df['Suhu_Udara'].mean()
plt.axvline(mean_suhu, color='red', linestyle='dashed', linewidth=1, label='mean')
plt.text(mean_suhu+1, 20, f'mean: {mean_suhu:.2f}', color='red')

# Tambahkan garis vertikal untuk menunjukkan nilai standar deviasi
std_suhu = df['Suhu_Udara'].std()
plt.axvline(mean_suhu+std_suhu, color='green', linestyle='dashed', linewidth=1, label='Std Dev')
plt.axvline(mean_suhu+std_suhu, color='green', linestyle='dashed', linewidth=1)
plt.text(mean_suhu+std_suhu+1, 15, f'Std Dev: {std_curah_hujan:.2f}', color='green')

plt.legend()
plt.show()

# Visualisasi dengan Boxplot
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.boxplot(df['Kelembapan_Udara'])
plt.title('Boxplot Kelembaban Udara')

plt.subplot(1, 2, 2)
plt.boxplot(df['Curah_Hujan'])
plt.title('Boxplot Curah Udara')

plt.tight_layout()
plt.show()

# Visulisasi dengan Scatterplot
plt.figure(figsize=(6, 6))
plt.scatter(df['Kelembapan_Udara'], df['Curah_Hujan'])
plt.xlabel('Kelembaban Udara')
plt.ylabel('curah Hujan')
plt.title('Scatterplot Kelembaban Udara vs cuarah hujan')
plt.show()
```
